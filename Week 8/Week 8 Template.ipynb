```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
apple = pd.read_csv("AAPL.csv")
apple['Date'] = pd.to_datetime(apple['Date'])
apple
```

```python
plt.figure(figsize=(10,5))
plt.plot(apple['Date'], apple['Close'])
plt.title("Apple Stock Closing Price")
plt.xlabel("Years")
plt.ylabel("Stock Price($)")
```

```python
# Add Trendline
y = apple['Close'].to_list()
x = [i for i in range(len(apple))]
coeffs = np.polyfit(x, y, 6)
y = np.polyval(coeffs, x)
?np.polyfit
```

```python
plt.close('all')
plt.figure(figsize=(10,5))
plt.plot(apple['Date'], apple['Close'])
plt.plot(apple['Date'], y)
plt.title("Apple Stock Closing Price")
plt.xlabel("Years")
plt.ylabel("Stock Price($)")
```

```python
# Box plot using Seaborn
import seaborn as sns
apple['Date_dow'] = apple['Date'].dt.dayofweek

fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
sns.boxplot(x="Date_dow", y='Close', data = apple)
ax.set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
ax.set(xlabel = 'Days of the Week', ylabel = 'Close Price ($)')
ax.set_title('Apple Box Plot of Stock Close Price by Day of the Week')
fig.savefig('box.png', dpi=400, transparent = True)
```

```python
# Violin plot using Seaborn

fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
sns.violinplot(x="Date_dow", y='Close', data = apple)
ax.set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])
ax.set(xlabel = 'Days of the Week', ylabel = 'Close Price ($)')
ax.set_title('Apple Violin Plot of Stock Close Price by Day of the Week')
fig.savefig('violin.png', dpi=400, transparent = True)
```

```python
# Convert close to dict
close = apple['Close'].to_list()
date = apple['Date'].dt.strftime('%m-%d')
date = date.to_list()
date_dict = dict()
for i in range(len(date)):
    if date[i] not in date_dict:
        date_dict.update({date[i]:[]})
    date_dict[date[i]].append(close[i])
    

date_dict = dict(sorted(date_dict.items()))
date_dict = {k : sum(v)/len(v) for k,v in date_dict.items()}
len(date_dict)
```

```python
# Create empty nested list
import numpy as np
date_nested_list = [[np.nan for _ in range(31)] for _ in range(12)]
# len(date_nested_list[0])
```

```python
def dateStrip(str_date: str) -> (int, int):
    str_date = str_date.split('-', 1)
    return [int(i) for i in str_date]
```

```python
date_list = list(date_dict.items())
date_list = [list(i) for i in date_list]
date_list = [[dateStrip(i[0]), i[1]] for i in date_list]
date_list
```

```python
for i in date_list:
    date_nested_list[i[0][0]-1][i[0][1]-1] = i[1]
    
date_nested_list
```

```python
import seaborn as sns
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
sns.heatmap(date_nested_list, xticklabels = [i if i % 2 == 1 else None for i in range(1, 32)], yticklabels = [i for i in range(1, 13)], cmap = 'autumn', annot = True)
ax.set(xlabel = 'Days', ylabel = 'Months')
ax.set_title('Apple Average Stock Close Price by Date')
fig.savefig('avg_price.png', dpi=400, transparent = True)
```

```python
# Exercise: Create a heatmap of apple['Close'] of year 2019
# If the date does not have a data, set it to np.nan
date = apple['Date'].to_list()
close = apple['Close'].to_list()
date_nested_list_19 = np.array([[np.nan for _ in range(31)] for _ in range(12)])
for d in range(len(date)):
    if date[d].year == 2019:
        date_nested_list_19[date[d].month - 1][date[d].day - 1] = close[d]
        
fig, ax = plt.subplots()
fig.set_size_inches(10, 5)
sns.heatmap(date_nested_list_19, xticklabels = [i if i % 2 == 1 else None for i in range(1, 32)], yticklabels = [i for i in range(1, 13)], cmap = 'autumn')
ax.set(xlabel = 'Days', ylabel = 'Months')
ax.set_title('Apple 2019 Stock Close Price by Date')
```
